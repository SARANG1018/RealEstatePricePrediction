## Train-Test Splitting

```python

# For learning purpose
import numpy as np
def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    
    print(shuffled)
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

## Scikit-learn Design

Primarily, three types of objects
1. Estimators - imputers
2. Transformers - fit() and fit_transform()
3. Predictors - LinearRegression model - fit(),predict(),score()

## Feature Scaling

Primarily, two types of feature scaling methods:
1. Min-max scaling (Normalization)
    (value - min)/(max - min)
    Sklearn provides a class called MinmaxScaler for this
    
2. Standardization
    (value - mean)/std
    Sklearn provides a class called Standard Scaler for this

## Implementation Of LLM ChatBot

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# Load pre-trained LLM model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
# Load your real estate price prediction model
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
import xgboost as xgb

from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
import numpy as np

base_models = [
    ('random_forest', GradientBoostingRegressor()),
    ('gradient_boosting', AdaBoostRegressor()),
    ('xgboost', xgb.XGBRegressor())
]

stacking_model = StackingRegressor(
    estimators=base_models,
    final_estimator=Lasso()  
)

price_prediction_model =stacking_model
```

```python
# Function to generate a response using the LLM model
def generate_response(question, document):
    # Tokenize question and document
    question_tokens = tokenizer.encode(question, add_special_tokens=False, return_tensors="pt")
    document_tokens = tokenizer.encode(document, add_special_tokens=False, return_tensors="pt")

    # Combine tokens
    input_tokens = torch.cat((question_tokens, document_tokens), dim=1)

    # Generate response from LLM model
    response_tokens = model.generate(input_tokens)

    # Decode the response
    response = tokenizer.decode(response_tokens[0], skip_special_tokens=True)

    return response
```

```python
# Fit the stacking model with training data
stacking_model.fit(housing_num_tr,housing_labels )

# Function to estimate the price using the price prediction model
def estimate_price(property_data):
    # Preprocess the property data using your pipeline
    property_prepared = my_pipeline.transform([property_data])

    # Use price prediction model to estimate the price
    predicted_price = stacking_model.predict(property_prepared)

    return predicted_price

user_question = "What is the average price for properties in this area?"
document_text = "This is a document containing real estate information."

response = generate_response(user_question, document_text)

new_property_data = [-0.43942006, 3.12628155, -1.12165014, -0.27288841, -1.42262747, -23.3846605, -1.31238772, 2.61111401, -1.0016859, -0.5778192, -0.97491834, 0.41164221, -0.86091034]
predicted_price = estimate_price(new_property_data)

print("Response:", response)
print("Predicted Price:", predicted_price)
```
